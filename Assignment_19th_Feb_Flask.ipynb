

**Q1. What is Flask Framework? What are the advantages of Flask Framework?**

Ans: Flask is a microframework for Python web development. It is a lightweight framework that is easy to learn and use. Flask is also very flexible, which makes it a good choice for a variety of web applications.

Some of the advantages of Flask include:

* **Easy to learn and use:** Flask is a very lightweight framework, which makes it easy to learn and use. Even beginners can get started with Flask quickly.
* **Flexible:** Flask is very flexible, which makes it a good choice for a variety of web applications. You can use Flask to create simple web applications or complex web applications.
* **Extendable:** Flask is extensible, which means that you can add new features to it using third-party extensions. This makes it easy to customize Flask to your needs.

**Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in Jupyter Notebook.**

Ans: here is the code to create a simple Flask application to display `Hello World!!`:

```python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "Hello World!!"

if __name__ == "__main__":
    app.run(debug=True)
```

To run this code, you can save it as a `.py` file and then run it from the command line. For example, if you save the code as `hello_world.py`, you can run it by typing the following command into the command line:

```
python hello_world.py
```

This will start the Flask application and open a web browser window to the `/` route. The `/` route is the default route in Flask, and it is the route that is used when you visit the application in a web browser.

**Q3. What is App routing in Flask? Why do we use app routes?**

Ans: App routing in Flask is the process of mapping URLs to functions in your application. This allows you to create different pages in your application that can be accessed by different URLs.

We use app routes because it allows us to organize our application code and make it easier to maintain. It also allows us to create dynamic URLs that can be used to access different pages in our application.
