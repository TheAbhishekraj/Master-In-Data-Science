##### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data  plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

```python
#### Answer :
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.figure(figsize=(6,2))
plt.scatter(x,y , c = '#223BDC',alpha=.9)
plt.xlabel("this is my x axis")
plt.ylabel("this is my y axis")
plt.title("this is my x vs y")
```

##### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])


```python
# Answer:

import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
x = company
y = profit

#Bar plot

plt.figure(figsize=(5,3))
plt.bar(x,y,color = '#7E10D5')
plt.xlabel("representing my profit data")
plt.ylabel("representing my company data")
plt.title("this is my bar plot")
plt.grid()
```

```python
#H Bar plot
plt.figure(figsize=(5,3))
plt.barh(x,y,color = '#7E10D5')
plt.xlabel("representing my profit data")
plt.ylabel("representing my company data")
plt.title("this is my bar plot")
plt.grid()
```

##### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

```python
#### Answer : 
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1,box2]
plt.figure(figsize=(10,7))
plt.boxplot(data)

plt.xlabel("representing my  data")
plt.ylabel("representing my range data")
plt.title("this is my box plot")

plt.show()
```
