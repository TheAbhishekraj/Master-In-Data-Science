**Q3. What is multiple inheritance? Write a python code to demonstrate multiple inheritance.**

Ans: Multiple inheritance is a feature of object-oriented programming that allows a class to inherit from multiple parent classes. This means that a class can inherit the properties and methods of multiple parent classes.

Here is an example of multiple inheritance in Python:

```python
class Vehicle:
  def __init__(self, name_of_vehicle, max_speed, average_of_vehicle):
    self.name_of_vehicle = name_of_vehicle
    self.max_speed = max_speed
    self.average_of_vehicle = average_of_vehicle

class Car(Vehicle):
  def seating_capacity(self, capacity):
    return f"The name of the vehicle is {self.name_of_vehicle} and its seating capacity is {capacity}"

class Truck(Vehicle):
  def carrying_capacity(self, capacity):
    return f"The name of the vehicle is {self.name_of_vehicle} and its carrying capacity is {capacity}"

class Bus(Car, Truck):
  def __init__(self, name_of_vehicle, max_speed, average_of_vehicle, capacity):
    super().__init__(name_of_vehicle, max_speed, average_of_vehicle)
    self.capacity = capacity

def main():
  bus = Bus("Bus", 120, 40, 50)
  print(bus.seating_capacity(50))
  print(bus.carrying_capacity(5000))

if __name__ == "__main__":
  main()
```

This code creates a `Bus` class that inherits from both the `Car` class and the `Truck` class. The `Bus` class then defines a method called `__init__` that takes four arguments: `name_of_vehicle`, `max_speed`, `average_of_vehicle`, and `capacity`. The `__init__` method then assigns these arguments to the instance variables of the `Bus` class.

The `Bus` class also defines two methods: `seating_capacity` and `carrying_capacity`. The `seating_capacity` method returns the name of the vehicle and its seating capacity, while the `carrying_capacity` method returns the name of the vehicle and its carrying capacity.

The `main` function in the code creates a `Bus` object and then prints the results of calling the `seating_capacity` and `carrying_capacity` methods.

In the above code, we have a base class `Vehicle` with a `start_engine` method that prints "Starting the engine". The `Car` class is a subclass of `Vehicle` and overrides the `start_engine` method with its own implementation, which prints "Starting the car engine".

When we create an object of the `Vehicle` class and call the `start_engine` method, the version of the method defined in the `Vehicle` class is executed. However, when we create an object of the `Car` class and call the `start_engine` method, the overridden version defined in the `Car` class is executed instead.

This demonstrates method overriding, where the subclass provides its own implementation of a method defined in the superclass, allowing for customization and specialization of behavior.
