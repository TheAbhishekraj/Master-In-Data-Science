Q5. Is Python code compiled or interpreted?
Ans: Python code is compiled into bytecode, which is then interpreted by the Python interpreter. When you run a Python program, the source code is first compiled into bytecode, which is a lower-level representation of the code. This bytecode is then executed by the Python interpreter. Hence, Python is often referred to as an interpreted language. However, it is important to note that Python uses a combination of compilation and interpretation.

Q8. Mention functions that can help us find the version of Python that we are currently working on.
Ans: There are multiple ways to find the version of Python you are currently working on:
- sys module: You can use the `sys` module and access the `sys.version` attribute to get the version information. For example:
```python
import sys
print(sys.version)
```
- platform module: The `platform` module provides cross-platform utilities and information, including the Python version. You can use `platform.python_version()` to retrieve the Python version. For example:
```python
import platform
print(platform.python_version())
```

Q14. How is memory managed in Python?
Ans: Memory management in Python is handled by the Python memory manager, which performs automatic memory allocation and deallocation. Python uses a combination of reference counting and a garbage collector for memory management.

- Reference counting: Python keeps track of the number of references to an object. When an object's reference count reaches zero, meaning it is no longer referenced by any part of the code

, the memory occupied by that object is automatically freed.

- Garbage collection: In addition to reference counting, Python also employs a garbage collector to reclaim memory from objects with circular references. Circular references occur when objects refer to each other, creating a reference cycle. The garbage collector periodically identifies and collects these cycles, freeing up memory.

Python's memory management allows developers to focus on writing code without worrying about explicit memory allocation and deallocation. The memory manager takes care of most memory-related tasks, resulting in simplified memory management for developers.

Q15. How to Install Python on Windows and set path variables?

2. Run the installer: Locate the downloaded installer file and double-click on it to launch the Python installer. Ensure that the "Add Python to PATH" option is checked during the installation process.

3. Customize the installation (optional): If you need to customize the installation, you can select the desired features and specify the installation directory. For a standard installation, you can proceed with the default settings.

4. Complete the installation: Click on the "Install Now" button to start the installation. The installer will copy the necessary files and set up Python on your system.

5. Verify the installation: After the installation is complete, open the command prompt and type "python" (without quotes) to check if Python is installed correctly. You should see the Python version information displayed.

6. Set up path variables (optional): To use Python from any location in the command prompt, you can add Python's installation directory to the system's PATH environment variable. Here's how:
   - Open the Start menu and search for "Environment Variables."
   - Select "Edit the system environment variables."
   - In the System Properties window, click on the "Environment Variables" button.
   - In the "System variables" section, select the "Path" variable and click on the "Edit" button.
   - Add the path to the Python installation directory (e.g., `C:\Python39`) to the list of paths. Separate multiple paths with a semicolon (;).
   - Click "OK" to save the changes.

Python is now installed on your Windows system, and you can use it by running the `python` command in the command prompt or by executing Python scripts.

Q16. Is indentation required in Python?
Ans: Yes, indentation is required in Python and plays a crucial role in the structure and execution of the code. In Python, indentation is used to define blocks of code, such as loops, functions, and conditional statements. Indentation helps to indicate the beginning and end of these blocks.

Unlike other programming languages that use braces or keywords to define blocks, Python uses indentation to enforce code structure. Indentation in Python is typically done using spaces or tabs (usually four spaces). Consistent and correct indentation is essential for Python code to execute correctly. If the indentation is incorrect or inconsistent, it can result in syntax errors or logical errors in the code.
