

**Q1. Explain GET and POST methods.**

Ans: The GET and POST methods are two of the most common HTTP methods used to interact with web applications. The GET method is used to retrieve data from a web application, while the POST method is used to send data to a web application.

The main difference between the GET and POST methods is that the GET method is used to retrieve data that is visible to the user, while the POST method is used to send data that is not visible to the user.

For example, if you want to retrieve the list of all users in a web application, you would use the GET method. However, if you want to create a new user in a web application, you would use the POST method.

**Q2. Why is request used in Flask?**

Ans:The `request` object in Flask is used to access the HTTP request that is being made to the application. The `request` object contains information about the request, such as the request method, the request headers, and the request body.

We use the `request` object in Flask to access the data that is being sent to the application by the user. For example, if the user submits a form in a web application, we can use the `request` object to access the data that was submitted in the form.

**Q3. Why is redirect() used in Flask?**

Ans: The `redirect()` function in Flask is used to redirect the user to a different URL. The `redirect()` function takes a URL as an argument and redirects the user to that URL.

We use the `redirect()` function in Flask to redirect the user to a different page in the application, or to a different application. For example, if the user clicks on a link to a page that does not exist, we can use the `redirect()` function to redirect the user to the home page of the application.

**Q4. What are templates in Flask? Why is the render_template() function used?**

Ans:Templates in Flask are used to render HTML pages. The `render_template()` function in Flask is used to render a template.

The `render_template()` function takes a template name and a dictionary of variables as arguments. The `render_template()` function returns an HTML page that is rendered with the variables that were passed to the function.

We use templates in Flask to make it easier to create and maintain HTML pages. Templates can be reused, and they can be easily updated.

**Q5. Create a simple API. Use Postman to test it. Attach the screenshot of the output in the Jupyter Notebook.**

Ans: here is the code to create a simple API:

```python
from flask import Flask

app = Flask(__name__)

@app.route("/api/hello")
def hello_world():
    return "Hello, World!"

if __name__ == "__main__":
    app.run(debug=True)
```

To test this API, we can use Postman. In Postman, we can create a new request and set the method to `GET`. The URL for the request should be `http://localhost:5000/api/hello`.

If we send the request, Postman will return the response `Hello, World!`.
