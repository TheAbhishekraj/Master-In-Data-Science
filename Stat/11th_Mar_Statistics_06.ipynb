##### Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5 using Python. Interpret the results.

```python
##### Ans:

import pandas as pd
from scipy import stats
import math
import matplotlib.pyplot as plt
import numpy as np

interval = 0.95
mean_val = 50
stdev = 5
n = 1
test_stat = stats.norm.ppf((interval + 1)/2)
lower_bound = mean_val - test_stat * stdev / math.sqrt(n)
upper_bound = mean_val + test_stat * stdev / math.sqrt(n)

print(test_stat,lower_bound,upper_bound)
```

##### Q3. Use Python to calculate the chi-square statistic and p-value for a contingency table with the following data:
Interpret the results of the test.
|       |Group A|Group B|
|-------|-------|-------|
|Outcome 1| 20| 15|
|Outcome 2| 10 |25|
|Outcome 3 |15 |20|

```python
##### Ans: 

from scipy.stats import chi2_contingency

# defining the table
data = [[20, 10, 15], [15, 25, 20]]
stat, p, dof, expected = chi2_contingency(data)
print(stat,p,dof)
# interpret p-value
alpha = 0.05
print("p value is " + str(p))
if p <= alpha:
	print('Dependent (reject H0)')
else:
	print('Independent (H0 holds true)')
```

##### Q4.
 A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the population who smoke.


```python
#### Ans:
import scipy.stats as stat
import numpy as np
import matplotlib.pyplot as plt
observed_data = [60,440] #Smoke & nOn smoke
import math

## find the critical value
significance=0.05
dof=len(observed_data)-1
critical_value=stat.chi2.ppf(1-significance,dof)
print(critical_value)

interval = 0.95
p = 0.12 # Smoke(60/500)
n = 500
#satandard Error of Population Priporation = se
se = np.sqrt((p * (1 - p))/n)
print(se)

# The t-multiplier is calculated based on the degrees of freedom and desired confidence level. 
# For samples with more than 30 observations and a confidence level of 95%, the t-multiplier is 1.96
# image.png
test_stat = stats.norm.ppf((interval + 1)/2)
lower_bound = p - test_stat * se
upper_bound = p + test_stat * se

print(test_stat,lower_bound,upper_bound)
```

##### Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the area corresponding to a chi-square statistic of 15.

```python
##### Ans:


from scipy.stats import chi2
import numpy as np
import matplotlib.pyplot as plt


chi2_value = 15
df = 10
# using the pdf function:
print(chi2.pdf(chi2_value, df))
# using the cdf function:
print(chi2.cdf(chi2_value, df))
chi2.sf(chi2_value, df)
x = np.arange(0, 20, 0.001)
leftTail = np.arange(0, chi2_value, 0.001)
plt.plot(x, chi2.pdf(x, df=df))
plt.fill_between(leftTail, chi2.pdf(leftTail, df=df), color='red')
plt.show()

```

```python
##### Q7. A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population who prefer Coke.
```

```python
#### Ans:
import scipy.stats as stat
import numpy as np
import matplotlib.pyplot as plt
observed_data = [520,480] #Coke or Pespi
import math

## find the critical value
significance=0.01
dof=len(observed_data)-1
critical_value=stat.chi2.ppf(1-significance,dof)
print(critical_value)

interval = 0.99
p = 0.52 # Coke(520/1000)
n = 1000
#satandard Error of Population Priporation = se
se = np.sqrt((p * (1 - p))/n)
print(se)

# The t-multiplier is calculated based on the degrees of freedom and desired confidence level. 
# For samples with more than 30 observations and a confidence level of 95%, the t-multiplier is 1.96
# image.png
test_stat = stats.norm.ppf((interval + 1)/2)
lower_bound = p - test_stat * se
upper_bound = p + test_stat * se

print(test_stat,lower_bound,upper_bound)


# Create Plot
# Create data
x_pdf = np.linspace(0.3, 1, 10000)
mean = p
std = se
y_pdf = stat.norm.pdf(x_pdf, mean, std)
# Create plot axis
ax = plt.axes()
ax.plot(x_pdf, y_pdf)
ax.set_title('Probability Distribution Function')
ax.set_ylabel('Relative Likelihood')
ax.set_xlabel('Value')
ax.set_ylim(0, np.max(y_pdf) * 1.05)
ax.set_xlim(0.4,0.6)
# Lower bound standard deviataion 
y = stat.norm.pdf(lower_bound, mean, std)
ax.vlines(lower_bound, 0, y, colors='k', ls='dotted')
plt.text(lower_bound, y * 1.05, r'$\mu - \sigma$', ha='right')
# Mean
y = stat.norm.pdf(mean, mean, std)
ax.vlines(mean, 0, y, colors='k', ls='dashed')
plt.text(mean, y * 1.02, r'Mean, $\mu$', ha='center')
# Plus one standard deviation
y = stat.norm.pdf(upper_bound, mean, std)
ax.vlines(upper_bound, 0, y, colors='k', ls='dotted')
plt.text(mean + std, y * 1.05, r'$\mu + \sigma$', ha='left')
#Fill color
plt.fill_between(x_pdf, y_pdf, 0,where = (x_pdf > lower_bound) & (x_pdf <= upper_bound),color = 'g')
plt.fill_between(x_pdf, y_pdf, 0,where =  (x_pdf < lower_bound),color = 'r')
plt.fill_between(x_pdf, y_pdf, 0,where =  (x_pdf > upper_bound),color = 'r')

# Finish
plt.show()
plt.close()

```
